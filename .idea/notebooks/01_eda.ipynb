```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# 1. Load the Data
# Note: We use '..' to go up one folder level to find 'data'
print("Loading data...")
classes = pd.read_csv('../data/raw/elliptic_txs_classes.csv')
features = pd.read_csv('../data/raw/elliptic_txs_features.csv', header=None)

# 2. Rename Columns for clarity
classes['class'] = classes['class'].map({'1': 'Illicit', '2': 'Licit', 'unknown': 'Unknown'})
print("Data Loaded.")

# 3. Visualize Class Imbalance
# This is the "Money Shot" for your problem statement
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=classes, palette='viridis')
plt.title('Distribution of Transactions (The Imbalance Problem)')
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.show()

# 4. Show the Numbers
print("Exact Counts:")
print(classes['class'].value_counts())
```
